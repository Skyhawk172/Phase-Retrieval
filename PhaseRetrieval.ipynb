```python
def zernikel(j, rho, phi):
    """
    Calculate Zernike polynomial with Noll coordinate j given a grid of radial
    coordinates rho and azimuthal coordinates phi.
    
    >>> zernikel(0, 0.12345, 0.231)
    1.0
    >>> zernikel(1, 0.12345, 0.231)
    0.028264010304937772
    >>> zernikel(6, 0.12345, 0.231)
    0.0012019069816780774
    """
    n = 0
    while (j > n):
        n += 1
        j -= n
    
    m = -n+2*j
    return zernike(m, n, rho, phi)


def zernike(m, n, rho, phi):
    """
    Calculate Zernike polynomial (m, n) given a grid of radial
    coordinates rho and azimuthal coordinates phi.
    """
    if (m > 0): return zernike_rad(m, n, rho) * N.cos(m * phi)
    if (m < 0): return zernike_rad(m, n, rho) * N.sin(m * phi)
    return zernike_rad(0, n, rho)


def zernike_rad(m, n, rho):
    """
    Calculate the radial component of Zernike polynomial (m, n)
    given a grid of radial coordinates rho.
    """
    wf = N.zeros(rho.shape)
    if (n-m) % 2:
        return wf
    for k in xrange((n-m)/2+1):
        wf += rho**(n-2.0*k) * (-1.0)**k * fac(n-k) / ( fac(k) * fac( (n+m)/2.0 - k ) * fac( (n-m)/2.0 - k ) )
        return wf

    
def zernike2(dim,n,m):

    y = np.arange(-dim/2,dim/2)
    x = np.arange(-dim/2,dim/2)
    xx,yy=np.meshgrid(x,y,sparse=True)

    rho = np.sqrt(xx**2 + yy**2)
    th= np.arccos(np.transpose(xx*1./rho))
    th= np.where(th<2.*np.pi,th,0)
    th= np.where(xx<0,2.*np.pi-th,th)
    

    if(type(n) is not int):
        raise Exception, "n must be integer"
    if(type(m) is not int):
        raise Exception, "m must be integer"
    if (n-m)%2!=0:
        raise Exception, "n-m must be even"
    if abs(m)>n:
        raise Exception, "The following must be true |m|<=n"
    mask=np.where(rho<=1,False,True)
    
    if(n==0 and m==0):
        return  np.ones( (dim,dim) )#masked_array(data=ones(rho.shape), mask=mask)
    rho=np.where(rho<0,0,rho)
    Rnm=np.zeros(rho.shape)

    S=(n-abs(m))/2

    for s in range (0,S+1):
        CR=pow(-1,s)*fac(n-s)/ (fac(s)*fac(-s+(n+np.abs(m))/2)* fac(-s+(n-np.abs(m))/2))
        p=CR*pow(rho,n-2*s)
        Rnm += p


    NC=np.sqrt(2*(n+1))
    S=(n-abs(m))/2
    
    if m>0:
        Zmn=NC*Rnm*np.cos(m*th)
    
    elif m<0:
        Zmn=NC*Rnm*np.sin(m*th)
    else:
        Zmn=np.sqrt(0.5)*NC*Rnm


    return Zmn/Zmn.max()

def create_pupil(dim,rad):
    pupil=np.zeros( (dim,dim) )
    i,j=np.meshgrid( np.arange(dim),np.arange(dim) )
    # circles contains the squared distance to the (100, 100) point
    # we are just using the circle equation learnt at school

# donuts contains 1's and 0's organized in a donut shape
# you apply 2 thresholds on circle to define the shape

    dist=np.sqrt( (i-dim/2)**2 + (j-dim/2)**2 )
    pupil[dist<rad/2]=1

    return pupil


def FFT(z):
    return np.fft.ifftshift(np.fft.fft2(np.fft.ifftshift(z)))

def IFFT(z):
    return np.fft.fftshift(np.fft.ifft2(np.fft.fftshift(z)))

def imshow(ifig,ipanel,string,img,size):
    plt.figure(ifig)
    mid=img.shape[0]/2
    plt.subplot(3,3,ipanel,axisbg='0.75', aspect="equal")
    plt.imshow( img )#,cmap=pl.cm.Greys_r) 
    plt.xlim(mid-size/2,mid+size/2)
    plt.ylim(mid-size/2,mid+size/2)
    plt.title(string)
    plt.show()

def resize(img,size):
    mid=img.shape[0]/2
    pad=size/2-mid
    return np.pad(img,(pad,pad),mode='constant')
```

```python
Nzer=10 #N zernike in test surface
Nsize=256

grid = (N.indices((Nsize, Nsize), dtype=N.float) - Nsize/2) / (Nsize/2)
grid_rho = (grid**2.0).sum(0)**0.5
grid_phi = N.arctan2(grid[0], grid[1])

grid_mask = grid_rho <= 1
#grid_mask = poppy.HexagonAperture().sample(npix=Nsize)

zern_list = [zernikel(i, grid_rho , grid_phi) * grid_mask for i in xrange(25)]

test_vec = N.random.random(Nzer)
test_vec = [test_vec[i] for i in xrange(len(test_vec))]
test_surf = sum(val * zernikel(i, grid_rho, grid_phi)/(i+1) for (i, val) in enumerate(test_vec))
guess_surf= sum(val * zernikel(i, grid_rho, grid_phi)/(i+1) for (i, val) in enumerate(test_vec))
noise = N.random.normal(0,0.05,(Nsize,Nsize))
test_surf += noise

### Try to reconstruct test surface

# Calculate covariance between all Zernike polynomials
cov_mat = N.array([[N.sum(zerni * zernj) for zerni in zern_list] for zernj in zern_list])

# Invert covariance matrix using SVD
cov_mat_in = N.linalg.pinv(cov_mat)

# Calculate the inner product of each Zernike mode with the test surface
wf_zern_inprod = N.array([N.sum(test_surf * zerni) for zerni in zern_list])

# Given the inner product vector of the test wavefront with Zernike basis,
# calculate the Zernike polynomial coefficients
rec_wf_pow = N.dot(cov_mat_in, wf_zern_inprod)
rec_wf_pow[rec_wf_pow<1e-12]=0.0

# Reconstruct surface from Zernike components
rec_wf = sum(val * zernikel(i, grid_rho, grid_phi) for (i, val) in enumerate(rec_wf_pow))



fig = plt.figure(1)
ax = fig.add_subplot(111)

ax.set_xlabel('Zernike mode')
ax.set_ylabel('Power [AU]')
ax.set_title('Reconstruction quality')

ax.plot(test_vec, 'r-', label='Input')
ax.plot(rec_wf_pow, 'b--', label='Recovered')
ax.legend()
#fig.savefig('py102-example2-plot1.pdf')

fig = plt.figure(3)
fig.clf()
ax = fig.add_subplot(111)
surf_pl = ax.imshow(test_surf* grid_mask, interpolation='nearest')
title("Test surface")
fig.colorbar(surf_pl)
#fig.savefig('py102-example2-plot2.pdf')
#fig.savefig('py102-example2-plot2.png')
#fig.savefig('py102-example2-plot2.eps')
```

```python
dim=Nsize
lambda0 = 2.12
xx=np.arange(0,dim,1)
yy=np.arange(0,dim,1)

WFE=1#000e-9
###################
# CIRCULAR PUPIL 
###################
pup_circ=create_pupil(dim,dim)
defocus =zernike2(dim,2,0)*grid_mask 
defocus =defocus/np.max(defocus)

test_surf = test_surf*grid_mask/np.max(test_surf)  #OPD units in microns 
test_phase= test_surf*2*np.pi/lambda0              #Phase map
opd_circ=test_surf

#pup_circ=ndimage.interpolation.zoom(pup_circ,4)
#opd_circ=ndimage.interpolation.zoom(opd_circ,4)

#################################
# RESIZE IMAGES
#################################
pupil0 = resize(pup_circ, 4*dim)
opd    = resize(opd_circ, 4*dim)
defoc  = resize(defocus,  4*dim)

scale_surf=resize(test_surf, 4*dim)

fig = plt.figure(3)
fig.clf()
ax = fig.add_subplot(111)
vmin=0.95*np.min(test_phase)
vmax=0.95*np.max(test_phase)
surf_pl = ax.imshow(test_phase, interpolation='nearest', vmin=vmin, vmax=vmax)
title("Test Phase Surface")
fig.colorbar(surf_pl)
```

```python
#################################
# CREATE DEFOCUSED IMAGES
#################################
phi=2.*np.pi*opd/lambda0
pupil_field=pupil0*np.exp(1.j*phi)
image_field= FFT(pupil_field)

PVtmp   = (np.max(defoc) - np.min(defoc))/lambda0 #physical units, wavelengths
Plus4   = 12*defoc/PVtmp 
Minus4  =-12*defoc/PVtmp
diversity=np.array([np.zeros( (4*dim,4*dim) ), Plus4, Minus4])

nimages    = 3
image0     = np.absolute(image_field) # np.sqrt(a**2 + b**2)
ImgDefocP4 = np.absolute( FFT(pupil_field*np.exp( 1j*2*np.pi* Plus4 / lambda0 ) ) ) 
ImgDefocM4 = np.absolute( FFT(pupil_field*np.exp( 1j*2*np.pi* Minus4/ lambda0 ) ) ) 

phase_est = pupil0*(0.2*np.random.random(image_field.shape) + scale_surf*2*np.pi/lambda0 )

#imshow(1,1,"",ImgDefocP4,dim/2)
#imshow(1,1,"",ImgDefocM4,dim/2)
```

```python
##########################
# GERCHBERG-SAXTON LOOP
##########################
amplitude0 = np.array([image0, ImgDefocP4, ImgDefocM4])
phase0    = np.array([phase_est, phase_est, phase_est])
for n in xrange(nimages):
    print "image %d," %n,
    for i in xrange(5): 
        image_field = FFT( pupil0*np.exp(1j*phase0[n]) * np.exp(1j*2*np.pi* diversity[n] / lambda0) )
        phase0[n]   = np.angle(image_field) 
        pupil_field = IFFT( amplitude0[n]*np.exp(1j*phase0[n]) ) *np.exp(-1j*2*np.pi* diversity[n] / lambda0)
        phase0[n]   = np.angle(pupil_field)
```

```python
phase = np.mean( phase0 , axis=0 )
mid=phase.shape[0]/2
final_phase = phase[mid-Nsize/2:mid+Nsize/2, mid-Nsize/2:mid+Nsize/2]

fig = plt.figure(3)
ax = fig.add_subplot(111)
surf_pl = ax.imshow(final_phase*pup_circ, interpolation='nearest', vmin=vmin, vmax=vmax)
title('Retrieved Phase')
fig.colorbar(surf_pl)
```

```python
### Try to reconstruct test surface

# Calculate the inner product of each Zernike mode with the test surface
wf_zern_inprod = N.array([N.sum(final_phase * zerni) for zerni in zern_list])

# Given the inner product vector of the test wavefront with Zernike basis,
# calculate the Zernike polynomial coefficients
rec_wf_pow = N.dot(cov_mat_in, wf_zern_inprod)
#rec_wf_pow[rec_wf_pow<1e-12]=0.0

# Reconstruct surface from Zernike components
rec_wf = sum(val * zernikel(i, grid_rho, grid_phi) for (i, val) in enumerate(rec_wf_pow))

pupil0_256=pupil0[mid-Nsize/2:mid+Nsize/2, mid-Nsize/2:mid+Nsize/2]
fig = plt.figure(3)
ax = fig.add_subplot(111)
surf_pl = ax.imshow(rec_wf*pupil0_256, interpolation='nearest', vmin=vmin, vmax=vmax)
title('Zernike Fit to Retrieved Phase')
fig.colorbar(surf_pl)

error = np.abs(final_phase-test_phase)

fig = plt.figure(4)
ax = fig.add_subplot(111)
surf_pl = ax.imshow(pupil0_256*error, interpolation='nearest')
title('Error b/w Input and Retrieved Phases')
fig.colorbar(surf_pl)
```
